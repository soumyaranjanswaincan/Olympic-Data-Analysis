OLYMPIC DATA ANALYSIS

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset 

athlete_events = pd.read_csv("/Users/jelenaraonic/UofTBootcamp/PROJECT1/project1/Olympic Data Analysis/Raw Data/athlete_events.csv")
regions = pd.read_csv("/Users/jelenaraonic/UofTBootcamp/PROJECT1/project1/Olympic Data Analysis/Raw Data/noc_regions.csv")
athlete_events.head()
regions.head()
```

```python
#join the data frame

athlete_df = athlete_events.merge(regions,how = "left",on = "NOC")
athlete_df.head()
```

```python
# print last 5 rows

athlete_df.tail(5)
```

```python
Countries_top10= athlete_df.Team.value_counts().sort_values(ascending= False).head(10)
Countries_top10
```

```python
plt.figure(figsize = (8,4))
plt.title('Countries with highest participation over the years')
sns.barplot(x = Countries_top10.index,y = Countries_top10,palette='Set1');
```

```python
#Gold metal athletes

gold_medals = athlete_df[(athlete_df.Medal == 'Gold')]
gold_medals.head()
```
